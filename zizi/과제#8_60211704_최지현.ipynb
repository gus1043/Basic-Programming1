```python
import collections
from operator import attrgetter

def chkname(string):
    if string == "":
        return False
    for i in string:
        if (i in "+-*/"):
            return False
    return True
    
def chknum(num):
    for i in num:
        if not (i in "0123456789"):
            return False
    return True

cnt=int(input("Input loop count: "))
print("-------------------------")
Student = collections.namedtuple("Student", "name stdid grade")
list=[]
while cnt>0:
    while True:
        name1 = input("input name: ")
        if chkname(name1):
            break
        print("이름을 잘못 입력하였습니다. 다시 입력하세요.")
    while True:
        stdid1 = input("input student id:")
        if chknum(stdid1):
            stdid1 = int(stdid1)
            break
        print("학번을 잘못 입력하였습니다. 다시 입력하세요.")
    while True:
        grade1 = input("input grade:")
        if chknum(grade1):
            grade1 = int(grade1)
            break
        print("학년을 잘못 입력하였습니다. 다시 입력하세요.")
    cnt-=1
    std = Student(name=name1, stdid=stdid1, grade=grade1)
    list.append(std)
    print("-------------------------")

list=sorted(list,key=attrgetter('stdid'))
print(list)
```

```python
def getDiscriminant(a, b, c):
    return (b ** 2) - (4 * a * c)

def getQuadraticFormula(a, b, c):
    QF = None

    if a == 0:
        if b == 0:
            # Answer is infinity
            if c == 0:
                return True

            # Answer is none
            else:
                return False

        # Answer is one real root
        else:
            return c / (-b)

    else:
        D = getDiscriminant(a, b, c)

    # Answer are two real root
    if D < 0:
        QF = (
            (-b + D ** 0.5) / (2 * a),
            (-b - D ** 0.5) / (2 * a))

    # Answer are two imaginary root
    elif D > 0:
        QF = (
            complex((-b) / (2 * a), (-D) ** 0.5 / (2 * a)),
            complex((-b) / (2 * a), - ((-D) ** 0.5) / (2 * a)))

    # Answer is multiple root
    else:
        QF = (-b) / (2 * a)

    return QF

print("Calculate QF(Quadratic Formula) for PSE(Perfect Square Expression)")
print("ex) ax^2 + bx + c = 0")

a = int(input("Input a: "))
b = int(input("Input b: "))
c = int(input("Input c: "))
QF = getQuadraticFormula(a, b, c)

print("Calculated answer is(are):")

if QF == True:
    print("Infinity.")

elif QF == False:
    print("None.")

elif isinstance(QF, tuple):
    for answer in QF:
        print("{:.2f}".format(answer))

else:
    print("{:.2f}".format(QF))
```
