```python
#실습1 제곱근 순차탐색--12
x=int(input('input number: '))
epsilon=0.01
step=epsilon
ans=0.0
print('currnet square root is ',ans)
while (abs(ans**2 - x) >= epsilon and ans <= x ): 
    ans += step
    if (ans**2 <= x) :
        print('currnet square root is ',round(ans,3))
    else :
        break
if abs(ans**2-x) >= epsilon:
    print ('Failed on square root of', x)
else :
    print(x, '\'s square root is', round(ans,3))
```

```python
#실습문제2
x=int(input('Input number: '))
epsilon=0.01
low = 0.0
high=max(1.0,x)
x_sqrt=(high+low)/2.0
while abs(x_sqrt**2 - x) >= epsilon:
    print('low=', low, 'high=', high, 'x_sqrt=', x_sqrt)
    if x_sqrt**2 < x:   
        low=x_sqrt
    else:
        high=x_sqrt
    x_sqrt=(high+low)/2.0
print(str(x)+'\'s square root is', x_sqrt)
```
