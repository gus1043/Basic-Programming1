```python
import random

class Building (object):

    def __init__(self):
        #Input the number of floors and the number of customers
        #Generate an elevator intance and customers
        
        while(1):
            num_of_floors = input("input num_of_floors:")

            if (num_of_floors.isdigit()):
                self.num_of_floors = int(num_of_floors)
                break
            print("Wrong Input!!!")

        while(1):
            num_of_customers = input("input num_of_customers:")

            if (num_of_customers.isdigit()):
                self.num_of_customers = int(num_of_customers)
                print("\n")
                break
            print("Wrong Input!!!")

        #Generate an Elevator instance
        self.elevator = Elevator(self.num_of_floors)

        #Generate custormers
        self.customer_list=[]
        for i in range(self.num_of_customers):
            self.customer_list.append(Customer(i+1, self.num_of_floors))

    def run(self):
    #run (loop) the elevator until every custormers reach (finish) their destinations.
    #For each floor, register the newly incoming customers, cancel the outgoing customers, 
    #and print the status of elevator and each customer. 
    #Once every customers reach their destination floors, exit the loop.    
    
        while(1):
            count_finished = 0

            self.elevator.register_customer(self.customer_list)
            self.elevator.cancel_customer(self.customer_list)
            
            self.output()

            print("\n->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->\n")

            for i in range(self.num_of_customers):
                if(self.customer_list[i].finished == True):
                    count_finished += 1
            if(count_finished == self.num_of_customers):
                print("\nAll Customers Arrived at Destination")
                break

            self.elevator.move()

    def output(self):
    #Print the current floor and the direction of the elevator, and status of each customer.
        print("\nElevator's Current Floor :", self.elevator.cur_floor, "\tElevator's direction:", self.elevator.direction)
        print("Customers: ")
        for i in range(len(self.customer_list)):
            print("\tCustomer", self.customer_list[i].ID, "\tcurrent floor :", self.customer_list[i].cur_floor, "\tdestination floor :", self.customer_list[i].dst_floor, 
                "\n\t\t\tin Elavator :", self.customer_list[i].in_elevator, "\tis Finished :", self.customer_list[i].finished)

class Elevator (object):
    
    def __init__(self, num_of_floors):
    #set the number of floors for the elevator, the current floor, 
    #and the direction "up" or "down"
        self.direction = "up"
        self.num_of_floors = num_of_floors
        self.cur_floor = random.randint(1, num_of_floors)
        if (self.cur_floor == self.num_of_floors):
            self.direction = "down"

    def move(self):
    #move the elevator by 1 "up" or "down" depending on the direction and 
    #the current floor. Print the elavator status. Change the elevator direction 
    #from "up" to "down" if the current floor is equivalent to the top floor, 
    #from "down" to "up" if the current floor is equivalent to the 1st floor. 
        if (self.direction == "up"):
            if(self.cur_floor == self.num_of_floors):
                print("The Elavator is on the Top Floor Now")
                self.cur_floor -= 1
                print("Elavator :", self.cur_floor+1, "->", self.cur_floor)
                self.direction = "down"
            else:
                self.cur_floor += 1
                print("Elavator :", self.cur_floor-1, "->", self.cur_floor)
        else:
            if(self.cur_floor == 1):
                print("The Elavator is on the Bottom Floor Now")
                self.cur_floor += 1
                print("Elavator :", self.cur_floor-1, "->", self.cur_floor)
                self.direction = "up"
            else:
                self.cur_floor -= 1
                print("Elavator :", self.cur_floor+1, "->", self.cur_floor)

    def register_customer(self, customer_list):
    #register the newly incoming customers for the current floor, adjust the flags,
    #and prompt that the customers entered the elevator.
        for i in range(len(customer_list)):
            if(customer_list[i].finished == False and customer_list[i].in_elevator == False):
                if(customer_list[i].cur_floor == self.cur_floor):
                    customer_list[i].in_elevator = True
                    print("Customer", customer_list[i].ID, "-> Elavator")

    def cancel_customer(self, customer_list):
    #cancel the outgoing customers, adjust the flags, 
    #and prompt that the customers reached their destination floors.
        for i in range(len(customer_list)):
            if(customer_list[i].finished == False and customer_list[i].in_elevator == True):
                if(customer_list[i].dst_floor == self.cur_floor):
                    customer_list[i].in_elevator = False
                    customer_list[i].finished = True
                    print("Customer", customer_list[i].ID, "-> Destination Floor")


class Customer (object):
    
    def __init__(self, ID, num_of_floors):
    # Set the customer's ID. Randomly select the customer's current floor and the destination floor. 
    # Initialize the flags. The customers's destination floor shall be different from the customer's current floor.
        self.ID = ID
        self.cur_floor = random.randint(1, num_of_floors)
        while(1):
            self.dst_floor = random.randint(1, num_of_floors)
            if self.cur_floor != self.dst_floor:
                break
        self.in_elevator = False
        self.finished = False
        
        

#test code
test = Building()
test.run()
```
