```python
integers=[]
n=0

def my_max(integers):
    return max(integers)

def my_min(integers):
    return min(integers)

def my_sum(integers):
    return sum(integers)

def my_avg(integers):
    return sum(integers)/n

def first(integers):
    if not (integers[0] in "0123456789-"):
        return False
    else:
        if str(integers)=='-':
            return False
        else:
            for c in integers[1:]:
                if not (c in "0123456789"):
                    return False
    return True

while True:
    x=input("양정수를 입력하세요: ")
    if first(x):
        if int(x) >= 0:
            n+=1
            integers.append(int(x))
        else:
            break
    else : 
        print ("다시 입력하세요")

if not integers:
    print("입력된 양수값이 없습니다.")
else:
    print ("최댓값은 ",my_max(integers))
    print('최솟값은 ', my_min(integers))
    print ('합은 ',my_sum(integers))
    print('평균은', my_avg(integers))
```
